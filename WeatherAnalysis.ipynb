```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("testset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
plt.plot(df['datetime_utc'],df[' _tempm'])
plt.xlabel("Date")
plt.ylabel("Temprature")
plt.title("temprature over time")
plt.show()
```

```python
plt.plot(df['datetime_utc'],df[' _hum'])
plt.xlabel("Date")
plt.ylabel("Humidity")
plt.title("Humidity over time")
plt.show()
```

```python
plt.plot(df['datetime_utc'],df[' _pressurem'])
plt.xlabel("Date")
plt.ylabel("Pressure")
plt.title("Pressure over time")
plt.show()
```

```python
plt.plot(df['datetime_utc'],df[ ' _wspdm'])
plt.xlabel("Date")
plt.ylabel("wind Speed")
plt.title("Wind Speed over time")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[' _conds'].value_counts().plot(kind='bar')
plt.xlabel("Weather condition")
plt.ylabel("Count")
plt.title("Frequency Different weather conditions")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[' _conds'].value_counts().plot(kind='pie')
plt.xlabel("Weather condition")
plt.ylabel("Count")
plt.title("Frequency Different weather conditions")
plt.xticks(rotation=90)
plt.show()
```
